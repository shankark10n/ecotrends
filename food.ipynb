```python
tab = [json.loads(l.strip()) for l in open('2006nps.json').readlines()]
```

```python
opening_garbage_pattern = '^[ \[]?u\''
closing_garbage_pattern = '\'$'
too_many_spaces_pattern = '  +'
messy_period_pattern = '([a-z]+)\. ?([a-z]+)'
mpp = re.compile(messy_period_pattern)
trailing_period = '\.$'

words = [re.sub(opening_garbage_pattern, '', s) for s in tab[1]['text'].split(',')]
words = [re.sub(closing_garbage_pattern, '', s) for s in words]
#words = [' '.join(mpp.split(s)) for s in words]
words = list(chain.from_iterable([mpp.split(s) for s in words]))
words = [l.strip() for l in words if l]
words = [re.sub(too_many_spaces_pattern, ' ', s) for s in words]
words = [re.sub(trailing_period, '', s) for s in words]
words = Counter(words)
```
